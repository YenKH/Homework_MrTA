```python
class interestRate:
    def __init__(self, date, name, type, term, rate):
        self.date=date
        self.name=name
        self.type=type
        self.term=term
        self.rate=rate
    def info(self):
        row = f'{self.date.ljust(15)} {self.name.ljust(30)} {self.type.ljust(25)} {self.term.ljust(15)} {self.rate}'
        return row
    def info_xuatDLT(self):
        row = f'{self.date}, {self.name}, {self.type}, {self.term}, {self.rate},'
        return row
    def info_xuatDLC(self):
        row = [self.date, self.name, self.type, self.term, self.rate]
        return row
```

```python
class interestRateList:
    def __init__(self, ds):
        self.ds=ds
        
    def loaddata(self, path):
        '''Hàm load dữ liệu từ file'''
        f = open(path,'r', encoding='utf8')
        for line in f :
            items = line.split(',')
            ir=interestRate(items[0], items[1], items[2], items[3], items[4])
            self.ds.append(ir)
        f.close()
            
    def add_data_to_file(self,path):
        '''Hàm ghi thêm dữ liệu vào file'''
        f= open(path, 'a+', encoding='utf8')
        
    def add (self, ir):
            self.ds.append(ir)
                        
```

```python
from tkinter import *
from tkinter import simpledialog
import datetime
import tkinter.messagebox as mbox
import os
import csv


#Nhập danh sách gốc
ds1= interestRateList([])
#ds1.loaddata(r'C:\Users\Admin\OneDrive - National Economics University\Desktop\test_interest_rate.txt')

#Khởi tạo
root = Tk()
root.geometry('700x469')
root.resizable(0,0)
root.title('HỆ THỐNG QUẢN LÝ THÔNG TIN LÃI SUẤT TIỀN GỬI TẠI NHTM')
myfont=('Helvetica', 11)

def nhap_tay():
    '''Hàm dùng để active khu vực nhập dữ liệu'''
    txt_date['state']='normal'
    txt_name['state']='normal'
    opt_type['state']='normal'
    opt_term['state']='normal'
    txt_rate['state']='normal'
    btn_luu['state']='normal'

def thoat(event):
    '''Dùng để thoát khỏi vùng nhập dữ liệu'''
    txt_date['state']='disabled'
    txt_name['state']='disabled'
    opt_type['state']='disabled'
    opt_term['state']='disabled'
    txt_rate['state']='disabled'
    btn_luu['state']='disabled'

def luu():
    '''Hàm dùng để lưu thông tin lãi suất nhập tay'''
    global ds1
    date=txt_date.get()
    name=txt_name.get()
    type=selected_type.get()
    term=selected_term.get()
    rate=txt_rate.get()
    if date!="" and name!="" and type!="" and term!="" and rate!="":
        try:
            int (date[5:7])
            int (date[8:])
            float(rate)
            try:
                int (date[5:7])
                int (date[8:])
                if int (date[5:7])<=12 and int(date[5:7])>0 and int (date[8:])>0 and int (date[8:])<=31:
                    try: 
                        float(rate)
                        ir=interestRate(date, name, type, term, rate)
                        qs=f'Bạn muốn lưu thông tin: \nNgày tháng: {date}; \nTên ngân hàng: {name};\nLoại hình: {type}; \nKỳ hạn: {term};\nLãi suất: {rate}'
                        cf=mbox.askyesno("XÁC NHẬN DỮ LIỆU", qs)
                        if cf == True:
                            ds1.add(ir)
                            mbox.showinfo("Trạng thái", "Lưu dữ liệu thành công")
                            display()
                            txt_date.delete(0,END)
                            txt_name.delete(0,END)
                            selected_type.set('-')
                            selected_term.set('-')
                            txt_rate.delete(0, END)
                    except:
                        mbox.showerror("Lỗi", "Lãi suất cần là số")
                else:
                    mbox.showerror("Lỗi", "Ngày/tháng sai")
            except:
                mbox.showerror("Lỗi", "Nhập sai định dạng ngày tháng. \nĐịnh dạng: YYYY-MM-DD")
        except:
            mbox.showerror("Lỗi", "Lỗi nhập liệu")
    #else:
        #mbox.showinfo("Lỗi", "Chưa nhập đủ trường dữ liệu")
    
def nhap_file():
    txt_date['state']='disabled'
    txt_name['state']='disabled'
    opt_type['state']='disabled'
    opt_term['state']='disabled'
    txt_rate['state']='disabled'
    btn_luu['state']='disabled'
    
    try:
        path_loaddata= simpledialog.askstring("Nhập đường dẫn", "Vui lòng nhập đường dẫn tới File cần nhập",
                                              parent=function_nhapDL)[1:-1]
        ds1.loaddata(path_loaddata)
        mbox.showinfo("Trạng thái", "Nhập dữ liệu thành công")
        display()
    except:
        mbox.showerror("Lỗi", "Dữ liệu chưa được nhập")        
        
def sua():
    '''Hàm dùng để sửa dữ liệu'''
    txt_date['state']='normal'
    txt_name['state']='normal'
    opt_type['state']='normal'
    opt_term['state']='normal'
    txt_rate['state']='normal'
    btn_luu['state']='normal'
    try:
        ir = lbx_display.get(lbx_display.curselection())
        for i in range(len(ds1.ds)):
            if ir==ds1.ds[i].info():
                qs1=f'Ngày tháng: {ds1.ds[i].date}; \nTên ngân hàng: {ds1.ds[i].name};\nLoại hình: {ds1.ds[i].type}; \nKỳ hạn: {ds1.ds[i].term};\nLãi suất: {ds1.ds[i].rate}'
                cf1=mbox.askyesno("XÁC NHẬN SỬA DỮ LIỆU", qs1)
                if cf1==True:
                    txt_date.insert(0,ds1.ds[i].date)
                    txt_name.insert(0,ds1.ds[i].name)
                    selected_type.set(ds1.ds[i].type)
                    selected_term.set(ds1.ds[i].term)
                    txt_rate.insert(0,ds1.ds[i].rate)
                    del ds1.ds[i]
                    break      
    except:
        mbox.showinfo("Thông báo", "Vui lòng chọn dòng dữ liệu cần sửa.")

def xoa():
    '''Hàm dùng để xóa dữ liệu'''
    try:
        ir = lbx_display.get(lbx_display.curselection())
        for i in range(len(ds1.ds)):
            if ir==ds1.ds[i].info():
                qs1=f'Ngày tháng: {ds1.ds[i].date}; \nTên ngân hàng: {ds1.ds[i].name};\nLoại hình: {ds1.ds[i].type}; \nKỳ hạn: {ds1.ds[i].term};\nLãi suất: {ds1.ds[i].rate}'
                cf1=mbox.askyesno("XÁC NHẬN XÓA DỮ LIỆU", qs1)
                if cf1==True:
                    del ds1.ds[i]
                    mbox.showinfo("Trạng thái", "Xóa thành công")
                    display()
                    break
    except:
        mbox.showerror("Thông báo", "Vui lòng chọn dòng dữ liệu cần xóa.")

def xuatTxt():
    try:
        with open ('Test_IR_List.txt', 'w', encoding = 'utf8') as f: 
            for ir in ds1.ds:
                f.write("%s\n" % ir.info_xuatDLT())
                noi_luu=os.getcwd()
                ms=f'Xuất dữ liệu thành công.\n Nơi lưu file:\n {noi_luu}\Test_IR_List.txt'
            mbox.showinfo("Trạng thái", ms)
    except:
        mbox.showerror("Lỗi", "Chưa xuất dữ liệu")
    
def xuatCsv():
    try: 
        Header = ['Ngày tháng', 'Tên ngân hàng', 'Loại hình', 'Kỳ hạn', 'Lãi suất']
        rows=[]
        for ir in ds1.ds:
            rows.append(ir.info_xuatDLC())
        with open ('Test_IR_List.csv', 'w', encoding = 'utf8') as f:
            write = csv.writer(f) 
            write.writerow(Header)
            write.writerows(rows)
            noi_luu=os.getcwd()
            ms=f'Xuất dữ liệu thành công.\n Nơi lưu file:\n {noi_luu}\Test_IR_List.csv'
        mbox.showinfo("Trạng thái", ms)
    except:
        mbox.showerror("Lỗi", "Chưa xuất dữ liệu")
    #pass

def display():
    lbx_display.delete(0,END)
    for ir in range(len(ds1.ds)):
        lbx_display.insert(0, ds1.ds[ir].info())

#Các chức năng

lbFrame_function = LabelFrame(root, text = "CHỨC NĂNG", font = myfont, fg='blue')
lbFrame_function.pack(fill = BOTH, expand = True)

#Nhập dữ liệu
function_nhapDL = Menubutton(lbFrame_function, text = 'Nhập dữ liệu', font = myfont, fg='Navy')
function_nhapDL.grid(row = 0, column = 0)
function_nhapDL.menu = Menu(function_nhapDL)
function_nhapDL["menu"]=function_nhapDL.menu
function_nhapDL.menu.add_command(label = "Nhập tay", command = nhap_tay)
function_nhapDL.menu.add_command(label = "Nhập File", command = nhap_file)

#sửa thông tin

function_sua = Menubutton(lbFrame_function, text = 'Thay đổi thông tin', font = myfont, fg='Navy')
function_sua.grid(row = 0, column = 1)
function_sua.menu = Menu(function_sua)
function_sua["menu"]=function_sua.menu
function_sua.menu.add_command(label = "Sửa", command = sua)
function_sua.menu.add_command(label = "Xóa", command = xoa)

#xuất dữ liệu
function_xuat = Menubutton(lbFrame_function, text = 'Xuất dữ liệu', font = myfont, fg='Navy')
function_xuat.grid(row = 0, column = 2)
function_xuat.menu = Menu(function_xuat)
function_xuat["menu"]=function_xuat.menu
function_xuat.menu.add_command(label = "Xuất ra file TXT", command = xuatTxt)
function_xuat.menu.add_command(label = "Xuất ra file CSV", command = xuatCsv)        
        
#vùng nhập dữ liệu

lbFrame_input = LabelFrame(root, text = "VÙNG NHẬP DỮ LIỆU", font = myfont, fg='blue')
lbFrame_input.pack(fill = BOTH, expand = True)

#Ngày tháng
lb_date = Label (lbFrame_input, width=12, height = 2, text = 'Ngày tháng: \n(yyyy-mm-dd)', font = myfont, justify = LEFT)
lb_date.grid(row = 0, column = 0, padx = 5, pady = 5)
txt_date= Entry(lbFrame_input, state='disabled')
txt_date.grid(row = 0, column = 1, padx = 5, pady = 5)
txt_date.bind('<KeyRelease>')#, valid_date)

#Tên ngân hàng
lb_name = Label (lbFrame_input, width=12, text = 'Tên ngân hàng:', font = myfont)
lb_name.grid(row = 0, column = 2, padx = 5, pady = 5)
txt_name= Entry(lbFrame_input, width=48, state ='disabled')
txt_name.grid(row = 0, column = 3, padx = 5, pady = 5)
txt_name.bind('<KeyRelease>')#, valid_name)

#Loại hình
lb_type = Label (lbFrame_input, width=12, text = 'Loại hình:', font = myfont, justify = LEFT)
lb_type.grid(row = 1, column = 0, padx = 5, pady = 5)
lst_type=['Tiền gửi Online', 'Tiền gửi tại quầy']
selected_type = StringVar()
opt_type = OptionMenu(lbFrame_input, selected_type, *lst_type)
opt_type.grid(row = 1, column = 1, padx = 5, pady = 5)
opt_type['state']='disabled'
#Kỳ hạn
lb_term = Label (lbFrame_input, text = 'Kỳ hạn:', font = myfont)
lb_term.grid(row = 1, column = 2, padx = 5, pady = 5)
lst_term=['1 tháng', '3 tháng', '6 tháng', '9 tháng', '12 tháng', '15 tháng', '18 tháng', '24 tháng', '36 tháng']
selected_term = StringVar()
opt_term = OptionMenu(lbFrame_input, selected_term, *lst_term )
opt_term.grid(row = 1, column = 3, padx = 5, pady = 5)
opt_term['state'] = 'disabled'

#Lãi suất
lb_rate = Label (lbFrame_input, text = 'Lãi suất:', font = myfont)
lb_rate.grid(row = 2, column = 0, padx = 5, pady = 5)
txt_rate= Entry(lbFrame_input, state='disabled')
txt_rate.grid(row = 2, column = 1, padx = 5, pady = 5)
txt_rate.bind('<KeyRelease>')#, valid_rate)
#lưu
btn_luu= Button(lbFrame_input, text = 'Lưu', font= myfont, bg='Green', fg = 'White', padx=40,
                state='disabled',
                command = luu)
btn_luu.grid(row = 2, column = 2, padx = 5, pady = 5)
btn_luu.bind("<Button-1>", luu())

#Thoát vùng nhập
btn_thoat= Button(lbFrame_input, text = 'Dừng nhập dữ liệu', font= myfont, bg='Red', fg = 'Black')
btn_thoat.grid(row = 2, column = 3, padx=5, pady=5)
btn_thoat.bind("<Button-1>", thoat)

#khu vực hiển thị thông tin
lbFrame_display = LabelFrame(root, text = 'DANH MỤC LÃI SUẤT', height=15, font = myfont, fg='Blue')
lbFrame_display.pack(fill = BOTH, expand = True)
lb_header = Label(lbFrame_display,
                  text = f'Ngày tháng'.ljust(25) + 'Tên ngân hàng'.ljust(40) + 'Loại hình'.ljust(20) + 'Kỳ hạn'.ljust(15) + 'Lãi suất (%)/năm',
                 font = ('Helvetia', 12), fg='Navy')
lb_header.pack(fill=BOTH)
lbx_display= Listbox(lbFrame_display,height=10, font = ('Helvetia', 12), bg='White')
display()
lbx_display.pack(fill = BOTH, expand = True)


#hiển thị ngày tháng
photo_time = PhotoImage(file=r'green.png')
lb_time = Label(root, font = ('Helvetia', 10),fg='Blue', image=photo_time, compound = LEFT)
lb_time.pack(side = 'right')
lb_time['text'] = f'{datetime.date.today()}'

root.mainloop()
```
